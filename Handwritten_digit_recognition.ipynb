```python
import numpy as np
import pandas as pd
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, AveragePooling2D, Conv2D, Flatten
from tensorflow.keras import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
(x_train,y_train),(x_test,y_test) = datasets.mnist.load_data()
```

```python
input_shape=(28,28,1)
x_train=x_train.reshape((x_train.shape[0],28,28,1))
x_test=x_test.reshape((x_test.shape[0],28,28,1))
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=0.2,random_state=40)
```

```python
x_train=(x_train - x_train.mean())/x_train.std()
x_val=(x_val - x_val.mean())/x_val.std()
x_test-(x_test - x_test.mean())/x_test.std()
```

```python
y_train=keras.utils.to_categorical(y_train,num_classes=10)
y_val=keras.utils.to_categorical(y_val,num_classes=10)
y_test=keras.utils.to_categorical(y_test,num_classes=10)
```

```python
y_train.shape
```

```python
cnn_model=Sequential()
cnn_model.add(Conv2D(6,kernel_size=(5,5),strides=(1,1),activation="tanh",input_shape=input_shape))
cnn_model.add(AveragePooling2D(pool_size=(2,2),strides=(1,1)))
cnn_model.add(Conv2D(16,kernel_size=(5,5),strides=(1,1),activation="tanh",input_shape=input_shape))
cnn_model.add(AveragePooling2D(pool_size=(2,2),strides=(1,1)))
cnn_model.add(Conv2D(120,kernel_size=(5,5),strides=(1,1),activation="tanh",input_shape=input_shape))

cnn_model.add(Flatten())
cnn_model.add(Dense(84,activation="tanh"))
cnn_model.add(Dense(10,activation="softmax"))
```

```python
cnn_model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
cnn_model.fit(x_train,y_train,batch_size=30,epochs=10,validation_data=(x_val,y_val))
```

```python
score=cnn_model.evaluate(x_test,y_test)
```
